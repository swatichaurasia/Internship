#### Importing Standard scaler to standardise all the numeric variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(iris.drop('Species', axis =1))
```

```python
scaled_features = scaler.transform(iris.drop('Species', axis = 1))
```

```python
df_feat = pd.DataFrame(scaled_features, columns= iris.columns[:-1])
df_feat.head()
```

#### Importing train_test_split to split data in training and testing set

```python
from sklearn.model_selection import train_test_split
```

Choosing K Value


```python
accuracy_rate= []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knn, df_feat, iris['Species'],cv=10)
    accuracy_rate.append(score.mean())
```

```python
error_rate =[]
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knn, df_feat, iris['Species'],cv=10)
    error_rate.append(1-score.mean())
```

Plot of Accuracy rate vs K

```python
plt.figure(figsize=(10,9))
plt.plot(range(1,40), accuracy_rate, color='blue', linestyle= 'dashed', 
         marker= 'o',markerfacecolor= 'red',markersize=10)
plt.title('Accuracy rate vs K')
plt.xlabel('K')
plt.ylabel('Accuracy_rate')
```

Plot of Error_rate vs K 

```python
plt.figure(figsize=(10,9))
plt.plot(range(1,40), error_rate, color='blue', linestyle= 'dashed', 
         marker= 'o',markerfacecolor= 'red',markersize=10)
plt.title('Error_rate vs K')
plt.xlabel('K')
plt.ylabel('Error_rate')
```

```python
# Error rate increases drastically after k=36. Lets retrain the model and check the classification report
```

```python
# First a quick caomparison with k=1 
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('With k=1')
print('\n')
print(confusion_matrix(y_test,y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```

```python
# Now with k= 36
knn = KNeighborsClassifier(n_neighbors=36)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('With k=1')
print('\n')
print(confusion_matrix(y_test,y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```
